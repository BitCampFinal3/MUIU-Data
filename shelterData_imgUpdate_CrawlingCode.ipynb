```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import json
import os

global no_places, no_photos, no_menus, no_times
no_places = 0
no_photos = 0
no_menus = 0
no_times = 0

def crawl_shelter_info(shelter):
    global no_places, no_photos, no_menus, no_times
    
    data = shelter.copy()
    data["imgUrls"] = []

    webDriver = webdriver.Chrome()

    equp_nm = shelter.get('equp_nm', '')

    if equp_nm.endswith("운동장"):
        equp_nm = equp_nm[:-3]

    url = f"https://map.naver.com/p/search/{equp_nm}"
    webDriver.get(url)
    wait = WebDriverWait(webDriver, 5)

    try:
        iframe_element = wait.until(EC.visibility_of_element_located((By.ID, "entryIframe")))
        webDriver.switch_to.frame(iframe_element)
    except Exception as e:
        no_places += 1
        print(f"iframe 전환 실패: {e}")
        webDriver.quit()
        return data

    try:
        body_element = wait.until(EC.visibility_of_element_located((By.CSS_SELECTOR, "#app-root > div > div > div > div.CB8aP > div")))
        img_elements = body_element.find_elements(By.TAG_NAME, "img")
        data["imgUrls"] = [img.get_attribute("src") for img in img_elements if img.get_attribute("src")]
        for img_url in data["imgUrls"]:
            print("대피소 사진:", img_url)
    except Exception as e:
        no_photos += 1
        print("대피소 사진 크롤링 실패:", e)

    webDriver.quit()
    return data

def process_json_files(input_dir, output_dir):
    global no_places, no_photos, no_menus, no_times

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    file_results = []

    for file_name in os.listdir(input_dir):
        if file_name.endswith('.json') and 'shelter' in file_name:
            file_path = os.path.join(input_dir, file_name)
            with open(file_path, 'r', encoding='utf-8') as file:
                shelter_data = json.load(file)

            if 'DATA' in shelter_data:
                for shelter in shelter_data['DATA']:
                    info = crawl_shelter_info(shelter)
                    file_results.append(info)
            else:
                print(f"파일 구조 오류: {file_name} - 'DATA' 키가 없음")

    output_file = os.path.join(output_dir, 'shelterData_imgUpdate.json')
    with open(output_file, 'w', encoding='utf-8') as file:
        json.dump({"DESCRIPTION": shelter_data.get("DESCRIPTION", ""), "DATA": file_results}, file, ensure_ascii=False, indent=4)

input_dir = 'C:/Users/Myhome/Desktop/Crawling'
output_dir = 'C:/Users/Myhome/Desktop/Crawling'
process_json_files(input_dir, output_dir)
```
